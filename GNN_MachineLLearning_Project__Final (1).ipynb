```python
from torch_geometric.datasets import Amazon
from torch_geometric.transforms import GCNNorm

dataset = Amazon(root='data/Amazon', transform=GCNNorm(),name = 'Computers')
data = dataset[0]

print('Dataset:', dataset)
print('-----------------------------')
print('Number of graphs: ',len(dataset))
print('Number of nodes: ',data.x.shape[0])
print('Number of features: ',dataset.num_features)
print('Number o classes: ',dataset.num_classes)
print('Edges are directed: ',data.is_directed())
print('Graph has isolated nodes: ',data.has_isolated_nodes())
print('Graph has loops: ',data.has_self_loops())
```

```python
from torch_geometric.utils import degree
from collections import Counter
import matplotlib.pyplot as plt

degrees = degree(data.edge_index[0]).numpy()
numbers = Counter(degrees)
fig, ax = plt.subplots()
ax.set_xlabel('Node degree')
ax.set_ylabel('Number of nodes')
plt.bar(numbers.keys(), numbers.values())
```

```python
train_mask = range(round(13752*0.70))
val_mask = range(round(13752*0.70) , round(13752*0.80))
test_mask = range(round(13752*0.8), 13752)
```

```python
import torch

torch.manual_seed(1)

import torch.nn.functional as F
from torch_geometric.nn import GATv2Conv, GCNConv
from torch.nn import Linear, Dropout
```

```python
dim_in = dataset.num_features
dim_h1 = 32
dim_h2 = 16
dim_out= dataset.num_classes
heads = 8
```

```python
gat1 = GATv2Conv(dim_in, dim_h1, heads=8)
gat2 = GATv2Conv(dim_h1 * heads, dim_h2, heads=8)
gat3 = GATv2Conv(dim_h2 * heads, dim_out, heads=8)

optimizer = torch.optim.Adam([{'params': gat1.parameters()},
                              {'params': gat2.parameters()},
                              {'params': gat3.parameters()}], lr=0.001, weight_decay=0.01)

epochs = 180
train_losses = []
val_losses = []
val_accuracies = []

for epoch in range(epochs + 1):
    optimizer.zero_grad()

    h = gat1(data.x, data.edge_index)
    h = F.elu(h)

    h2 = gat2(h, data.edge_index)
    h2 = F.elu(h2)

    out = gat3(h2, data.edge_index)
    out = F.log_softmax(out, dim=1)

    train_loss = F.nll_loss(out[train_mask], data.y[train_mask])
    train_losses.append(train_loss.item())
    correct_train = out[train_mask].argmax(dim=1) == data.y[train_mask]
    train_acc = sum(correct_train) / len(correct_train)

    val_loss = F.nll_loss(out[val_mask], data.y[val_mask])
    val_losses.append(val_loss.item())
    correct_val = out[val_mask].argmax(dim=1) == data.y[val_mask]
    val_acc = sum(correct_val) / len(correct_val)
    val_accuracies.append(val_acc.item())

    loss = train_loss + 0.5 * val_loss

    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print('Epoch ',epoch,' || ','Train Acc: ',train_acc.item()*100,'%',' || ','Val Acc:', val_acc.item() * 100,'%')

correct_test = out[test_mask].argmax(dim=1) == data.y[test_mask]
predicted_labels = out.argmax(dim=1).numpy()
actual_labels = data.y[test_mask].numpy()
test_acc = sum(correct_test) / len(correct_test)
print('GAT test accuracy:', round(test_acc.item() * 100, 2))
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

with torch.no_grad():
    h = gat1(data.x, data.edge_index)
    h = F.elu(h)
    h2 = gat2(h, data.edge_index)
    h2 = F.elu(h2)
    embeddings = gat3(h2, data.edge_index)


tsne = TSNE(n_components=2, random_state=42)
embeddings_tsne = tsne.fit_transform(embeddings.cpu().numpy())


plt.figure(figsize=(10, 8))
colors = predicted_labels
scatter = plt.scatter(embeddings_tsne[:, 0], embeddings_tsne[:, 1], c=colors, cmap='viridis')
plt.title('t-SNE Visualization of GAT Predictions')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend(*scatter.legend_elements(), title='Predicted Classes')
plt.show()
```
